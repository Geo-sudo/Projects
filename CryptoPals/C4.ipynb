**Detect single-character XOR**

One of the 60-character strings in this file has been encrypted by single-character XOR.

Find it.

```python
from crypto_functions import hex_to_bytes, bytes_to_hex

with open("4.txt", "r") as file:
    byte_string_list = [hex_to_bytes(line.strip()) for line in file]
```

### Generating the Messages by XORing with Every Possible Key

```python
def generate_XORed_messages(byte_arr):
    key_list = {}
    for i in range(0, 256):
        decrypted_list = []

        for j in range(0, len(byte_arr)):
            XORed_byte = byte_arr[j]^i
            decrypted_list.append(chr(XORed_byte))

        decrypted_text = "".join(decrypted_list)
        key_list[i] = [decrypted_text, 0]
    return key_list    
```

### Scoring the Messages

```python
def score_text(string: str):
    score = 0
    frequency_list = [ ' ', 'e', 't', 'a', 'o', 'i', 'n', 's', 'r', 'h', 'd', 'l', 'u', 
    'c', 'm', 'f', 'y', 'w', 'g', 'p', 'b', 'v', 'k', 'x', 'q', 'j', 'z']

    for char in string.lower():
        if char in frequency_list:
            score += (len(frequency_list) - frequency_list.index(char))
        else:
            score-=30

    return score
```

```python
def score_text_list(key_list):
    for key, message in key_list.items():
        message_score = score_text(message[0])
        key_list[key][1] = message_score
    return key_list
```

### Ranking the Messages

```python
def rank_messages(key_list):
    best_message = None
    best_key = None
    best_score = float('-inf') 

    for key, message in key_list.items():
        current_key = key
        current_message = message[0]
        current_score = message[1]

        if current_score>best_score:
            best_key = current_key
            best_message = current_message
            best_score = current_score
    return best_key, best_message, best_score
```

### Finding the Message (by Running All Previous Functions)

```python
def find_best(byte_arr):
    key_list = generate_XORed_messages(byte_arr)
    key_list = score_text_list(key_list)
    return rank_messages(key_list)
```

### Getting a List of the Best (Highest Ranking) Messages of Each String in the TXT File

```python
best_list = []
for byte_arr in byte_string_list:
    key, message, score = find_best(byte_arr)
    best_list.append([key, message, score, byte_arr])
```

### Ranking the Messages (Finding the Best of the Best)

```python
def rank_messages_list(best_list):
    best_message = None
    best_key = None
    best_score = float('-inf') 
    best_byte_arr = ""
    for decrypted in best_list:
        current_key = decrypted[0]
        current_message = decrypted[1]
        current_score = decrypted[2]
        current_byte_arr = decrypted[3]

        if current_score>best_score:
            best_key = current_key
            best_message = current_message
            best_score = current_score
            best_byte_arr = current_byte_arr
    return best_key, best_message, best_score, best_byte_arr
```

```python
best_key, best_message, best_score, best_byte_arr = rank_messages_list(best_list)
best_hex_string = bytes_to_hex(best_byte_arr)
print(f"The key is {best_key}\nThe message '{best_message}'\nFrom the Hex string '{best_hex_string}'\nWith a score of {best_score}")
```
