**Detect AES in ECB mode**

In this file are a bunch of hex-encoded ciphertexts.

One of them has been encrypted with ECB.

Detect it.

Remember that the problem with ECB is that it is stateless and deterministic; the same 16 byte plaintext block will always produce the same 16 byte ciphertext.

```python
from crypto_functions import hex_to_bytes

with open("8.txt", "r") as file:
    hex_bytes_list = [hex_to_bytes(line.strip()) for line in file]
```

```python
def detect_AES_ECB(bytes_list: [bytearray]):
    ranked_bytes_list = []
    for bytes_line in bytes_list:
        chunk_counter = 0
        divided_chunks = []
        for i in range(0, len(bytes_line), 16):
            divided_chunks.append(bytes_line[i:(i+16)])

        for i in range(0, len(divided_chunks)-1):
            for j in range(i+1, len(divided_chunks)):
                if divided_chunks[i] == divided_chunks[j]:
                    chunk_counter += 1
        
        ranked_bytes_list.append([chunk_counter, bytes_line])
    
    return ranked_bytes_list
```

```python
def detect_AES_ECB(bytes_list: [bytearray]):
    best_score = -1
    detected_line = None

    for bytes_line in bytes_list:
        divided_chunks = []
        unique_chunks = set()
        for i in range(0, len(bytes_line), 16):
            divided_chunks.append(bytes_line[i:(i+16)])

        for chunk in divided_chunks:
            unique_chunks.add(bytes(chunk))
        
        score = len(divided_chunks) - len(unique_chunks)

        if score>best_score:
            best_score = score
            detected_line = bytes_line
    
    return best_score, detected_line
```

```python
ranked_AES = detect_AES_ECB(hex_bytes_list)
```
